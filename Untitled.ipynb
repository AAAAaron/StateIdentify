```python
Fs=50#hz
#ax,ay,az,anx,any,anz,wx,wy,wz,wnx,wny,wnz
def tensor_make(dfdata,dataname):
    freqlen=dfdata.shape[0]
    fq=np.linspace(0,Fs,freqlen)
    result=np.empty((freqlen,len(dataname),2))
    for index,item in enumerate(dataname):
#         print item
        data=dfdata[item]
        fftresult=fft(data,freqlen)       
        fm=abs(fftresult)
#         yf1=fm          #归一化处理
#         yf1[0]=0
#         yf2 = yf1[range(int(len(dfdata)/2))]  #由于对称性，只取一半区间
#         xf = np.arange(len(dfdata))*Fs/len(dfdata)        # 频率
#         xf2 = xf[range(int(len(dfdata)/2))]  #取一半区间
        fm[0]=fm[0]/freqlen
        fm[1:]=fm[1:]/freqlen*fm[0]#把模值复现出来
        
        fa=np.angle(fftresult)
        result[:,index,0]=fm
        result[:,index,1]=fa
        fm[0]=0
        feq=fq[fm.tolist().index(max(fm))]
        result[-1,index,0]=feq
    return result
    
```

```python
%%time
slide_windows_length=50
windows_last=100
Fs=50
total_length=sensordata.shape[0]
dataname=['加速度x','加速度y','加速度z','anx','any','anz','x轴旋转角','y轴旋转角','z轴旋转角','wnx','wny','wnz','全加速度模制滤波']
total_tensor=np.empty((int(total_length/slide_windows_length)+1,windows_last+1,len(dataname),2))
for index,item in enumerate(range(0,total_length-windows_last,slide_windows_length)):
    total_tensor[index]=tensor_make(sensordata.loc[item:item+windows_last,:],dataname)
```
